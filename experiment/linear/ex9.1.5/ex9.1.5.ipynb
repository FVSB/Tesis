# Importar las depencias necesarias

```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x)
BilevelJuMP.@variable(Lower(model),0<=y[1:2])
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-x+10y[1]-y[2])
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,-y[1]-y[2])
BilevelJuMP.@constraints(Lower(model),begin
c1,x-y[1]<=1
c2,x+y[2]<=1
c3,y[1]+y[2]<=1
c4,-y[1]<=0
c5,-y[2]<=0
end)
```

## Run

```julia
start_experiment(model,[x],y,"ex9.1.5(Linear)")
```
