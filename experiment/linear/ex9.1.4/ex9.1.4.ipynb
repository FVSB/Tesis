# Importar las depencias necesarias

```julia
include("../../experiment.jl")
```

```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),x)
BilevelJuMP.@variable(Lower(model),y)
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,x-4*y)
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,y)
BilevelJuMP.@constraints(Lower(model),begin
c1,-2*x+y<=0
c2,2*x+5*y-108<=0
c3,(2*x)-(3*y)+4<=0
end)
```

## Run

```julia
start_experiment(model,[x],[y],"ex9.1.2(Linear)")start_experiment(model,[x],[y],"ex9.1.4(Linear)")
```
