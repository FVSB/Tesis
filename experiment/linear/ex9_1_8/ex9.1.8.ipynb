# Importar las depencias necesarias

```julia
include("../../experiment.jl")
```

```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x[1:2])
BilevelJuMP.@variable(Lower(model),0<=y[1:2])
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-2x[1]+x[2]+0.5y[1])
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,x[1]+x[2]-4y[1]+y[2])
BilevelJuMP.@constraints(Lower(model),begin
c1,-2x[1]+y[1]-y[2]<=-2.5
c2,x[1]-3x[2]+y[2]<=2
c3,x[1]+x[2]<=2
end)
```

## Run

```julia
start_experiment(model,x,y,"ex9.1.9(Linear)")
```
