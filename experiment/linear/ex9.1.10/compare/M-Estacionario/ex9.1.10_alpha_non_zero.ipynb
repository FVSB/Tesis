```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x[1:2])
BilevelJuMP.@variable(Lower(model),0<=y[1:2])
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-2x[1]+x[2]+0.5y[1])
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,-4y[1]+y[2]+x[1]+x[2])
BilevelJuMP.@constraints(Lower(model),begin
c1,2.5 - 2x[1] + 1.3361855683301653y[1] - 0.5389628631323411y[2]<=0
c2,-2 + x[1] - 3x[2] - 0.00748967074829618y[1] + 0.9897288382275685y[2]<=0
c3,-0.5759991579683006y[1] + 0.5814649784365273y[2]<=0
c4,0.47606101920412214y[1] - 0.34714089495770184y[2]<=0
end)
```
