```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),x)
BilevelJuMP.@variable(Lower(model),0<=y[1:2])
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,0.5*((y[1]-2)^2+(y[2]-2)^2))
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,0.5y[1]^2+y[2])
```

```julia
BilevelJuMP.@constraints(Lower(model),begin
c1, -x + y[1] + y[2]==0    # Es de igualdad
c2,-7.51 + 1.26y[1] + 0.3y[2] <=0
c3,-3.99 + 0.13y[1] - 0.98y[2] <=0
end)
```

## Run

```julia
start_experiment(model,[x],y,"ex9.2.4(Quadratic)_alpha_non_zero")
```
