Definir variables

```julia
BilevelJuMP.@variable(Upper(model), 0<=x<=8)
BilevelJuMP.@variable(Lower(model), 0<=y[1:2]<=4)
```

### Nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-x-3y[1]+2y[2])
```

### Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,-y[1])
BilevelJuMP.@constraints(Lower(model),begin
c1,-y[1]<=0
c2,y[1]<=4
c3,-2x+y[1]+4y[2]<=16
c4,8x+3y[1]-2y[2]<=48
c5,-2x+y[1]-3y[2]<=-12

end)
```
