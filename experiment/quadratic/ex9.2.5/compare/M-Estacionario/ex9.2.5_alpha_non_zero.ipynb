# Importar las depencias necesarias

```julia
include("../../../../experiment.jl")
```

```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x<=8)
BilevelJuMP.@variable(Lower(model),y)
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,(x-3)^2+(y-2)^2)
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,(y-5)^2)
```

```julia
BilevelJuMP.@constraints(Lower(model),begin
c1,-9.75 + x + 1.23y<=0
c2,-9.77 + x + 1.24y<=0
c3,-5.07 - 2x + 3.58y<=0

end)
```

## Run

```julia
start_experiment(model,[x],[y],"ex9.2.5(Quadratic)_alpha_non_zero")
```
