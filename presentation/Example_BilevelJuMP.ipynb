# Importar las depencias necesarias

```julia
include("../experiment/experiment.jl")
```

```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x<=15)
BilevelJuMP.@variable(Lower(model),0<=y<=20)
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,x^2+(y-10)^2)
```

```julia
BilevelJuMP.@constraints(Upper(model),begin
a1,-x+y<=0
end)
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,(x+2y-30)^2)
BilevelJuMP.@constraints(Lower(model),begin
c1,x+y<=20
c2,-y<=0
c3,y<=20

end)
```

## Run

```julia
start_experiment(model,[x],[y],"Example")
```
