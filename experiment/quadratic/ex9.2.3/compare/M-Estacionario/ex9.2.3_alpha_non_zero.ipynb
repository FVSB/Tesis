```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x[1:2]<=50)
BilevelJuMP.@variable(Lower(model),-10<=y[1:2]<=20)
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,2x[1]+2x[2]-3y[1]-3y[2]-60)
```

```julia
BilevelJuMP.@constraints(Upper(model),begin
a1,x[1]+x[2]+y[1]-2y[2]-40<=0
end)
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,(y[1]-x[1]+20)^2+(y[2]-x[2]+20)^2)
```

```julia
BilevelJuMP.@constraints(Lower(model),begin
c1,10 - x[1] - 0.2404132953131409y[1] + 1.6558082001777785y[2]<=-0
c2,10 - x[2] - 0.585728900171587y[1] + 1.161431217456761y[2]<=-0
c3,-10 - 0.22932442500548145y[1] + 1.1033508479257956y[2]<=0
c4,-20 + 0.7319144150322708y[1] - 0.3838093058715799y[2]<=0
c5,-10 + 1.1364666153873229y[1] + 0.6270418375927904y[2]<=0
c6,-20 - 0.19666084308720935y[1] + 0.7184471455766357y[2]<=0
end)
```
