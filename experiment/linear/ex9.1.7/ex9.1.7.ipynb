```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x[1:2])
BilevelJuMP.@variable(Lower(model),0<=y[1:3])
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-8x[1]-4x[2]+4y[1]-40y[2]+4y[3])
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,x[1]+2x[2]+y[1]+y[2]+2y[3])
BilevelJuMP.@constraints(Lower(model),begin
c1,-y[1]+y[2]+y[3]<=1
c2,2x[1]-y[1]+2y[2]-0.5y[3]<=1
c3,2x[2]+2y[1]-y[2]-0.5y[3]<=1
c4,-y[1]<=0
c5,-y[2]<=0
c6,-y[3]<=0
end)
```
