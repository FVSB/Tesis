```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x[1:2]<=50)
BilevelJuMP.@variable(Lower(model),-10<=y[1:2]<=20)
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,2x[1]+2x[2]-3y[1]-3y[2]-60)
```

```julia
BilevelJuMP.@constraints(Upper(model),begin
a1,x[1]+x[2]+y[1]-2y[2]-40<=0
end)
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,(y[1]-x[1]+20)^2+(y[2]-x[2]+20)^2)
```

```julia
BilevelJuMP.@constraints(Lower(model),begin
c1,10 - x[1] - 0.1453425393127531y[1] + 0.695382405408262y[2]<=-0
c2,10 - x[2] - 0.540858965354142y[1] - 2.8548355198014805y[2]<=-0
c3,-10 - 0.7610620421579353y[1] + 2.144744857840972y[2]<=0
c4,-20 + 0.9971267947967347y[1] - 0.025790343823472256y[2]<=0
c5,-10 + 0.19010638513018685y[1] + 0.7064249465972601y[2]<=0
c6,-20 - 0.01714153002851966y[1] + 0.8461349183854041y[2]<=0
end)
```
