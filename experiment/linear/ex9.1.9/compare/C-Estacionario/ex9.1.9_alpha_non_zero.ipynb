# Importar las depencias necesarias

```julia
include("../../../../experiment.jl")
```

```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x)
BilevelJuMP.@variable(Lower(model),0<=y)
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,x+y)
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,-5x-y)
BilevelJuMP.@constraints(Lower(model),begin
c1,-1.89- x + 3.06y<=0
c2,1.42 - 0.25x<=0
c3,-5.54415 + x - 0.055y<=0
c4,-23.23 + x + 6.76y<=0
c5,-17.58 - 7.12y<=0

end)
```

## Run

```julia
start_experiment(model,[x],[y],"ex9.1.9(Linear)_alpha_non_zero")
```
