```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x[1:2])
BilevelJuMP.@variable(Lower(model),0<=y[1:2])
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-2x[1]+x[2]+0.5y[1])
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,-4y[1]+y[2]+x[1]+x[2])
BilevelJuMP.@constraints(Lower(model),begin
c1,1.75 - 2x[1] + y[1] - y[2]<=0
c2,6.55 + x[1] - 3x[2] + y[2]<=0
c3,104.6 - y[1]<=0
c4,-2.05 - y[2]<=0
end)
```

## Run

```julia
start_experiment(model,x,y,"ex9.1.10(Linear)_alpha_zero")
```
