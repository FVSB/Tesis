# Importar las depencias necesarias

```julia
include("../../experiment.jl")
```

```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

### Declarar las variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x)
BilevelJuMP.@variable(Lower(model),0<=y)
```

## Declarar el nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-x-3y)
```

### Delcrar el nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,y)
BilevelJuMP.@constraints(Lower(model),begin
c1,-y<=0
c2,-x+y<=3
c3,x+2y<=12
c4,4x-y<=12
end)
```

### Realizar experimentacion 

```julia
start_experiment(model,[x],[y],"ex9.1.2(Linear)")
```
