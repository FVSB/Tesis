```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model), -0.5<=x<=1)
BilevelJuMP.@variable(Lower(model), -0.5<=y<=1)
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,x)
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,-y^2)
```

## Run

```julia
start_experiment(model,[x],[y],"mitos_2(nonconvex)")
```
