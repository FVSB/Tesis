```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x[1:2])
BilevelJuMP.@variable(Lower(model),0<=y[1:2])
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-2x[1]+x[2]+0.5y[1])
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,-4y[1]+y[2]+x[1]+x[2])
BilevelJuMP.@constraints(Lower(model),begin
c1,2.5 - 2x[1] + 0.7642501908369801y[1] - 1.1662572942554035y[2]<=0
c2,-2 + x[1] - 3x[2] - 0.4709850940862368y[1] + 0.6678482682322896y[2]<=0
c3,-0.2062254575395409y[1] + 0.5597917900626571y[2]<=0
c4,0.4709850940862368y[1] - 0.6678482682322896y[2]<=0
end)
```

## Run
