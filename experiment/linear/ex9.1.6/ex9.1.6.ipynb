# Importar las depencias necesarias

```julia
include("../../experiment.jl")
```

```julia
using BilevelJuMP
```

```julia
model=BilevelModel()
```

## Definir variables

```julia
BilevelJuMP.@variable(Upper(model),0<=x)
BilevelJuMP.@variable(Lower(model),0<=y)
```

## Definir nivel superior

```julia
BilevelJuMP.@objective(Upper(model),Min,-x-3y)
```

## Definir Nivel inferior

```julia
BilevelJuMP.@objective(Lower(model),Min,-x+3y)
BilevelJuMP.@constraints(Lower(model),begin
c1,-x-2y<=-10
c2,x+2y<=6
c3,2x-y<=21
c4,x+2y<=38
c5,-x+2y<=18
c6,-y<=0
end)
```

## Run

```julia
start_experiment(model,[x],[y],"ex9.1.6(Linear)")
```
